```python
from typing import Any


predictionsCSV = pd.read_csv("../MTourneyPredictions.csv")
teamsCSV = pd.read_csv("../data/mens/MTeams.csv")

teamNames: dict[int, str] = dict[int, str](
    zip[tuple](teamsCSV["TeamID"].astype(int), teamsCSV["TeamName"].astype(str))
)

predictionPairs: set[tuple[int, int]] = set[tuple[int, int]](
    zip[tuple[int, int]](
        predictionsCSV["WTeamID"].astype(int),
        predictionsCSV["LTeamID"].astype(int),
    )
)
round1Results = []
round1Points = 0

for _, row in matchup1CSV.iterrows():
    winnerID = int(row["WinnerID"])
    loserID = int(row["LoserID"])

    if (winnerID, loserID) in predictionPairs:
        correct = True
        predictedWinnerID = winnerID
        pointsEarned = ROUND1INCR
    elif (loserID, winnerID) in predictionPairs:
        correct = False
        predictedWinnerID = loserID
        pointsEarned = 0
    else:
        correct = False
        predictedWinnerID = None
        pointsEarned = 0

    round1Points += pointsEarned
    round1Results.append({
        "Winner": teamNames[winnerID],
        "Loser": teamNames[loserID],
        "YourPick": teamNames[predictedWinnerID] if predictedWinnerID is not None else "Missing",
        "Correct": correct,
        "PointsEarned": pointsEarned,
    })

round1ResultsCSV = pd.DataFrame(round1Results)
display(round1ResultsCSV.head())
print(f"First round points: {round1Points}")
```
